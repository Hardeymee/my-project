## Data Cleaning
1. Handling the missing data
2. Data type conversion

```python
#dropping columns 
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df_clean = df.dropna()

old_shape = df.shape
new_shape = df_clean.shape

old_shape, new_shape
```

```python
df_clean.info()
```

```python
df_clean['seat'].value_counts()
```

```python
##replacing '5 or 7' with the mode - 5

df_clean['seat'].replace('5 or 7', '5', inplace = True)
```

```python
df_clean['seat'].unique()
```

```python
#converting the 'seat' column datatype
df_clean['seat'] = df_clean['seat'].astype('int64')
```

```python
df_clean.dtypes
```

```python
#split the 'region' column into two columns 'state' and 'city'
df_clean[['state', 'city']] = df_clean['region'].str.split(',', expand=True)
```
